```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import qmc

# 1. Define the sampler (Dimension = 2)
sampler = qmc.Sobol(d=2, scramble=True)

# 2. Generate points (usually a power of 2 for better distribution)
n_points = 256
sample = sampler.random(n=n_points)

# 3. Generate standard random points for comparison
random_sample = np.random.uniform(0, 1, (n_points, 2))

# 4. Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Sobol Plot
ax1.scatter(sample[:, 0], sample[:, 1], s=15, color='#2ca02c')
ax1.set_title(f'Sobol Sequence (n={n_points})')
ax1.set_aspect('equal')

# Random Plot
ax2.scatter(random_sample[:, 0], random_sample[:, 1], s=15, color='#d62728')
ax2.set_title(f'Standard Random (n={n_points})')
ax2.set_aspect('equal')

plt.show()
```
